```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Create dataset
data = pd.DataFrame({
    "text": [
        "Payment not processed",
        "Unable to login",
        "Need help with pricing",
        "App crashing",
        "Refund not received"
    ],
    "category": ["Billing", "Technical", "General", "Technical", "Billing"],
    "priority": ["High", "High", "Low", "High", "High"]
})

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data["text"])

# Train models
model_cat = MultinomialNB()
model_cat.fit(X, data["category"])

model_pri = MultinomialNB()
model_pri.fit(X, data["priority"])

# Test input
new_ticket = ["Payment failed"]
new_X = vectorizer.transform(new_ticket)

print("Category:", model_cat.predict(new_X)[0])
print("Priority:", model_pri.predict(new_X)[0])
test_tickets = [
    "Payment failed",
    "App is crashing",
    "Need help with account"
]

for ticket in test_tickets:
    vec = vectorizer.transform([ticket])
    print("Ticket:", ticket)
    print("Category:", model_cat.predict(vec)[0])
    print("Priority:", model_pri.predict(vec)[0])
    print("------")
    from sklearn.metrics import accuracy_score

# Predict on training data
cat_pred = model_cat.predict(X)
pri_pred = model_pri.predict(X)

print("Category Accuracy:", accuracy_score(data["category"], cat_pred))
print("Priority Accuracy:", accuracy_score(data["priority"], pri_pred))
```
